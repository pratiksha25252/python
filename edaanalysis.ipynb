```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
Data = {
    'exp': [2, 4, 5, 3, 1],
    'salary': [20000, 60000, 40000, 30000, 15000]
}

# Create DataFrame
df = pd.DataFrame(Data)
    
# Define X and y
X = df[['exp']]
y = df[['salary']]

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Predict
predicted_salary = model.predict(X)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='skyblue', label='Actual Data')
plt.plot(X, predicted_salary, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.title('Linear Regression: Experience vs Salary')
plt.legend()
plt.grid(True)
plt.show()

# Print model parameters
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data={
     'car': ['Sedan A', 'SUV B', 'Coupe C', 'Sedan A', 'SUV B', 'Coupe C', 'Sedan A'],
    'year':      [2018,      2019,     2020,     2021,      2022,     2023,     2024],
    'price':     [20000,     35000,    27000,    22000,     37000,    29000,    23000],
    'mileage':   [30000,     25000,    15000,    28000,     23000,    14000,    26000]
}

df=pd.DataFrame(data)
print(df['price'].mean())
avgprice=df.groupby('car')['price'].mean()
print(avgprice)
minmilege=df['mileage'].idxmin()
carmin=df.loc[minmilege,'car']
print(carmin)
plt.xlabel('Car Model')
plt.ylabel('Average Price')
plt.title('Average Price per Car Model')
plt.barh(avgprice.index,avgprice,color='blue')
plt.show()
plt.hist(df['mileage'],bins=20)
plt.show()
goodcars=(df[(df['price']<50000) & (df['mileage']<50000)])
print(goodcars)
plt.scatter(df['price'],df['mileage'],color='blue',label='good value cars',s=100)
plt.axhline(20000, color='red', linestyle='--', label='Price = $20,000')
plt.axvline(50000, color='blue', linestyle='--', label='Mileage = 50,000')
plt.grid(True)
plt.legend()



```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Create synthetic dataset
data = {
    'CustomerID': range(1, 11),
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ivy', 'Jack'],
    'Age': [25, 30, 22, 40, np.nan, 28, 35, 42, 200, 29],  # Notice the outlier 200 and NaN
    'Gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'F', 'F', 'M'],
    'City': ['New York', 'Chicago', 'Chicago', 'LA', 'New York', 'Chicago', 'LA', 'NYC', 'Chicago', 'New York'],
    'PurchaseAmount': [250, 400, 150, 800, 500, np.nan, 300, 2500, 270, 400]  # Notice outlier 2500 and NaN
}
df = pd.DataFrame(data)
print(df.info())
print(df.describe())
print(df.shape)
print("null values",df.isnull().sum())
print(df.duplicated())
df=df.drop_duplicates()
for col in ['Age','PurchaseAmount']:
    df[col].fillna(df[col].mean(),inplace=True)
print(df.isnull().sum())
numeric=['Age','PurchaseAmount']
for col in numeric:
    q1=df[col].quantile(.25)
    q2=df[col].quantile(.75)
    iqr=q2-q1
    lower=q1-1.5*iqr
    upper=q2+1.5*iqr
dfiqr=df[(df[col]>=lower) & df[col]<=upper]
print(dfiqr)
print(df["Age"].mean())
print(df["PurchaseAmount"].sum())
print(df.groupby('City')['PurchaseAmount'].mean())
plt.hist(df['Age'],bins=10,color='green')
sns.boxplot(x='Age',y='PurchaseAmount',data=df)
avg_purchase = df.groupby('Gender')['PurchaseAmount'].mean().reset_index()
plt.show()

sns.barplot(x='Gender', y='PurchaseAmount', data=avg_purchase, palette='pastel')
plt.title('Average Purchase Amount by Gender')
plt.ylabel('Average Purchase')
plt.xlabel('Gender')
plt.show()

corr=df[['Age','PurchaseAmount']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
city_counts = df['City'].value_counts()

plt.pie(city_counts, labels=city_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Customer Distribution by City')
plt.axis('equal')
plt.show()



```
